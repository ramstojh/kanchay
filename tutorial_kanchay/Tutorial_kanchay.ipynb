# There Is A k'anchay That Never Goes Out

## Downloading TESS light curves

```python
from kanchay import kan
%matplotlib inline
```

k'anchay uses the lightkuve code to find TESS light curves (LC) of a given star. The tool downloads and plots all the SPOC light curves observed by TESS in several sectors. The tool also normalizes and applies sigma clipping to the LCs. The resultant LC is stored in arrays in x (time), y (flux) and yerr (flux error).

```python
starid='Gaia DR2 2395082091638750848'
kan.search_lk(starid, mission='TESS')
```

k'anchay also stores both the unnormalized and normalized LC in a csv table. For that, just add store ='yes' to the above command. Then, a csv table will be created with the TIC (TESS Input Catalog) ID of the star. The table has several columns where '_n' is for the normalized LC (e.g, time_n, flux_n, err_flux_n).

```python
starid='Gaia DR2 2395082091638750848'
kan.search_lk(starid, mission='TESS', store='yes')
```

By default, k'anchay downloads SPOC TESS light curves, which means those with 2 minute cadence data. However, it is possible to download those with 20 seconds cadence.

```python
starid='HIP 1599' #γ Pav
kan.search_lk(starid, mission='TESS', exptime=20)
```

## Measuring Rotational Periods

If you are interested in measuring rotational periods, k'anchay will not disappoint you, since it is based on the powerful starspot and exoplanet codes. 
Run the search_lk module again, but now store your data in the x, y and yerr arrays.

```python
starid='Gaia DR2 2395082091638750848'
x, y, yerr = kan.search_lk(starid, mission='TESS')
```

Now, you can determine rotational periods with only one command (thanks to the starspot code!)
The rotational period is estimated using three methods (LS, ACF, PDM).

```python
kan.rotation_calc(x[0], y[0], yerr[0])
```

Nice results. All the methods give the same rotational period value (1.59 days).
You can also determine rotational periods using GP (thanks to the exoplanet code!) by adding gp='yes'. But sometimes it is better to cut the LC to get better rotational periods.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 5))
plt.scatter(x[0][2000:8500], y[0][2000:8500])
```

Now. let's compute the rotational period using GP

```python
kan.rotation_calc(x[0][2000:8500], y[0][2000:8500], yerr[0][2000:8500], gp='yes')
```

The rotational period estimated using GP is in complete agreement with those from LS, ACF and PDM.
